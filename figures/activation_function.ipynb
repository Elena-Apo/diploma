```python
import numpy as np
import matplotlib.pyplot as plt     # display images
#https://github.com/christianversloot/machine-learning-articles/blob/main/a-simple-conv3d-example-with-keras.md
```

```python
# Activation functions
def ReLU(x):
    return np.maximum(np.zeros_like(x), x)

def step(x):
    return x>0

def tanh(x):
    return (np.exp(x) - np.exp(-x))/(np.exp(x) + np.exp(-x))

def sigmoid(x):
    return 1/(1 + np.exp(-x))   
```

```python
x = np.linspace(-3,3,100)
y_ReLU = ReLU(x)

fig,ax = plt.subplots(figsize=(5.3,3))
plt.plot(x, y_ReLU,linewidth=2, markersize=12)
plt.rcParams.update({'font.size':14})

plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('$y(x) = \mathrm{ReLU}(x)$')
#Saving results
image_name = f'ReLU.jpg'
fig.savefig(image_name, format='jpg', dpi=150, bbox_inches='tight', pad_inches = 0.2) 
```

```python
x = np.linspace(-3,3,100)
y_step = step(x)

fig,ax = plt.subplots(figsize=(5.3,3))
plt.plot(x, y_step,linewidth=2, markersize=12)
plt.rcParams.update({'font.size':14})

plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('$y(x) = \mathrm{step}(x)$')
#Saving results
image_name = f'step.jpg'
fig.savefig(image_name, format='jpg', dpi=150, bbox_inches='tight', pad_inches = 0.2)
```

```python
x = np.linspace(-3,3,100)
y_step = tanh(x)

fig,ax = plt.subplots(figsize=(5.3,3))
plt.plot(x, y_step,linewidth=2, markersize=12)
plt.rcParams.update({'font.size':14})

plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('$y(x) = \mathrm{tanh}(x)$')
#Saving results
image_name = f'tanh.jpg'
fig.savefig(image_name, format='jpg', dpi=150, bbox_inches='tight', pad_inches = 0.2)
```

```python
x = np.linspace(-5,5,100)
y_step = sigmoid(x)

fig,ax = plt.subplots(figsize=(5.3,3))
plt.plot(x, y_step,linewidth=2, markersize=12)
plt.rcParams.update({'font.size':14})

plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('$y(x) = \mathrm{sigmoid}(x)$')
#Saving results
image_name = f'sigmoid.jpg'
fig.savefig(image_name, format='jpg', dpi=150, bbox_inches='tight', pad_inches = 0.2)
```
