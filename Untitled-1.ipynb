```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f_row = [679, 770, 852, 941]
f_s = 8000
a = 2*np.pi*f_row[0]/f_s

N = 8000

S1 = np.sin(a)
C1 = np.cos(a)
S = np.zeros(N)
C = np.zeros(N)

C[0] = 1

for n in range(1,N):
    S[n] = C1*S[n-1] + S1*C[n-1]
    C[n] = C1*C[n-1] - S1*S[n-1]

fig,ax = plt.subplots(figsize=(10,5), dpi=100)
# plt.subplot(121)
plt.plot(S[:50])
# plt.subplot(122)
plt.plot(C[:50])
```
